```python
# class 클래스 -> 객체를 표현하기 위한 문법의 일종
# 프로그래밍으로 객체를 만들 때 사용하는 것

# 전사, 마법사, 궁수, 자동차, 비행기 -> object
# -> 현실, 가상, 개념

# 속성(명사적) - attribute / 저장의 개념
# 능력(기능, 동사적) - method


class Person: # 설계도
    def hello(self): # self 매개변수 자리에 꼭 있어야 함
        print("hi")

jin = Person() # jin은 Person의 인스턴스(객체)
jin.hello()
```

```python
class Person():
    def __init__(self):
        self.hello = '안녕하세요' # 인스턴스 속성
    
    def hi(self):
        print(self.hello)

jin = Person() # 이때 __init__ method 이미 실행됨
jin.hi()
```

```python
class Person():
    def __init__(self, name, age, hobby):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby
    
    def introduce(self):
        print("{}, 저는 {}입니다.".format(self.hello, self.name))

jin = Person('진', 22, '운동')
jin.introduce()

print('이름:', jin.name)
print('나이:', jin.age)
print('취미:', jin.hobby)
```

```python
# 클래스 외부에서는 접근 불가, 내부에서만 사용 가능한 private attribute
class Person():
    def __init__(self, name, age, hobby, weight):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby
        self.__weight = weight

jin = Person('진', 22, '운동', 78)
jin.__weight -= 20

print(jin.__weight)
```

```python
class Person():
    def __init__(self, name, age, hobby, weight):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby
        self.__weight = weight

    def diet(self, loss_weight):
        self.__weight -= loss_weight
        print("몸무게 :", self.__weight)

jin = Person('진', 22, '운동', 78)
jin.diet(10)
```

```python
# class 속성
class Person():
    folder = [] # 클래스 속성 -> 클래스에 속함(공공제) 모든 인스턴스에서 접근 가능

    def add_files(self, file):
        self.folder.append(file)

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
    
```

```python
# 인스턴스 속성
class Person():
    def __init__(self):
        self.folder = []

    def add_files(self, file):
        self.folder.append(file)

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```
